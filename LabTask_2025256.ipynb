```python
#Task no 1:-
def generate_invoice(customer_name, *prices, tax_rate=0.05, **details):
    """
    In this function we take input the costumer name and a unpacked list of prices and
    it calculates the tax using a default parameter 0.05 and diplays the result
    in a dictionary.
    """
    subtotal = 0
    for price in prices:
        subtotal = subtotal + price

    tax = subtotal * tax_rate
    total = subtotal + tax
    
    return {
        "customer_name": customer_name,
        "subtotal": subtotal,
        "tax": tax,
        "total": total,
        "details": details
    }

#Function Calling
price_list = [100, 200, 300]
contact_info = {"email": "ABCD@gmail.com"}
shipping_info = {"city": "XYZ"}

print(generate_invoice(
    "ali bhai", 
    *price_list, 
    **contact_info, 
    **shipping_info
))

```

```python
#Task no 2:-
def safe_average(*numbers):
    """
    This function calculates the average of the numbers and returns returns none 
    if no arguments are given and invalid if something other than numbers are entered
    returns the average if corrct input is given.
    """
    if len(numbers) == 0:
        return "none"

    # Using map and lambda
    validation = list(map(lambda x: isinstance(x, (int, float)), numbers))
    
    if False in validation:
        return "Invalid Input"
        
    return sum(numbers) / len(numbers)

# Testing
print(f"Only numbers: {safe_average(10, 20, 30)}")
print(f"Mixed values: {safe_average(10, 'ABC', 30)}")
print(f"No arguments: {safe_average()}")
```
