```python
# comments
-> single inheritance, a child class inherits from a single-parent class
-> multiple inheritance, one child class can inherit from multiple parent classes
-> multilevel inheritance, a class inherits from a child class or derived class
-> Hierarchical inheritance, more than one child class is derived from a single parent class.
-> inheritance is consists of multiple types or a combination of different inheritance is called hybrid inheritance.
-> In child class, we can refer to parent class by using the super() function.
-> Polymorphism in Python is the ability of an object to take many forms.
    -> duck typing -> A function calls .fly() on an object. It doesnt care if the object is a Bird``Airplane, or Superman, as long as they all have a .fly()
    -> method overriding ->  feature that allows a subclass (child class) to provide a specific implementation of a method that is already defined in its superclass (parent class).
    -> method overloading ->The ability to define multiple methods within the same class that have the same name but different parameter lists (different number, types, or order of arguments).
    -> operator overloading ->Allowing built-in operators (like +``-``*``==) to be redefined to work with user-defined types (classes/structures).
    
```

```python
# single inhertance
class animal:
    def sound(self):
        print ("paretn")
class dog(animal):
    def bark(self):
        print("child")
c = dog()
c.bark()
c.sound()
```

```python
# multiple inheritacne
class insan:
    def sound(self):
        print ("parent one")
class hevan:
    def name(self):
        print ("second parent")
class dog(insan,hevan):
    def bark(self):
        print("insan")
c = dog()
c.bark()
c.sound()
c.name()
```

```python
# multiple inheritacne
class vehicle:
    def vehicle_info(self):
        print('this is vehicle')

class car(vehicle):
    def car_info(self,name):
        print("car name is ",name)

class truck(vehicle):
    def truck_info(self,name):
        print("name of truck is",name)

obj1=car()
obj1.vehicle_info
obj1.car_info
obj2=truck()
obj2.vehicle_info()
obj2.truck_info('ewa')
```

```python
#multilevel inheritance
class vehicle:
    def vehicle_info(self):
        print("Inside Vehicle Class")

class car(vehicle):
    def car_info(self):
        print("Inside car class")

class sports_car(car):
    def sports_info(self):
        print("Inside sports car class")
    
s_car=sports_car()
s_car.car_info()
s_car.vehicle_info()
s_car.sports_info()
```

```python
#Hierarchical inheritance
class Vehicle:
    def vehicle_info(self):
        print("This is a vehicle")

class Car(Vehicle):
    def car_info(self, name):
        print("Car name is:", name)

class Truck(Vehicle):
    def truck_info(self, name):
        print("Name of truck is:", name)

# Testing the Car object
obj1 = Car()
obj1.vehicle_info()      
obj1.car_info("Tesla")   

# Testing the Truck object
obj2 = Truck()
obj2.vehicle_info()
obj2.truck_info("Ewa")  
```

```python
#hybird inheritance
class vehicle:
    def vehicle_info(self):
        print('this is vehicle class')

class car(vehicle):
    def car_info(self):
        print("inside car class")

class truck(vehicle):
    def truck_info(self, name):
        print(f"inside truck class: {name}")
class sportscar(car):
    def sportscar_info(self):
        print("inside sports car class")
class sportsCar(car,truck):
    def sportscar_info(self):
        print("inside sportscar class")

# Creating an instance and calling methods
s_car = sportscar()
s_car.vehicle_info()  # Inherited from vehicle
s_car.car_info()      # Inherited from car
s_car.sportscar_info() # Defined in sportscar
obj1 = sportsCar()
obj1.truck_info("mazda")
```
