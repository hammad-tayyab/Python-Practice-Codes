```python
#Part (A) :
class FullTimeEmployee: #defining fulltime employee
    def salary(self):
        return 50000
    def work_type(self): 
        return "Working full-time"

class PartTimeEmployee: #defining part time employee
    def salary(self):
        return 20000
    def work_type(self):
        return "Working part-time"

employees = [FullTimeEmployee(), PartTimeEmployee()] # storing objeects in a list

for emp in employees: #polymorphism calls the required class
    if emp.salary() == 50000:
        print(f"Full Time employee salary: {emp.salary()}")
    else:
        print(f"Part Time employee salary: {emp.salary()}")
    
    print(emp.work_type(),"\n") 

#Part B;
class ProjectHours: #difining project hours class and it's member functions.
    def __init__(self, hours):
        self.hours = hours

    def __add__(self,others):
        return ProjectHours(self.hours + others.hours)
# Creating objects
project1 = ProjectHours(30)
project2 = ProjectHours(30)
# Adding
total = project1 + project2 # __add__ is called internally..
print(f"Total Project Hours: {total.hours}")    
```
