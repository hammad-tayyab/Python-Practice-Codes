```python
#TASK 1
string = input("Enter a string: ")
string = string.lower()
l1 = list(string)
keys = []
isSpace = False
dic = {}
st=""
ignore = ['and','the','is']
for i in range(len(string)):
    if l1[i] == ' ':
        isSpace = True
        keys.append(st)
        st = ""
    else:
        isSpace = False
        st = st + l1[i]
if st != ' ':
    keys.append(st)

for key in keys:
    if key not in ignore:
        if key in dic:
            dic[key] +=1
        else:
            dic[key] = 1
print (dic)        
```

```python
#TASK 2
d1 = {'a':{'x':1,'y':2},'b':[1,2],'c':3}
d2 = {'a':{'x':3,'z':4},'b':[3,4],'d':5}
d3 = {}
l1 = list(d1.keys())
l2 = list(d2.keys())
l3 = list(set(l1 + l2)) # i am using set to remove duplicates
for  i in l3:
    if i in d1 and i in d2:
        if type(d1[i]) == dict:
            dict1 = list(d1[i].keys())
            dict2 = list(d2[i].keys())
            dict3 = list(set(dict1 + dict2)) # using set to remove duplictes
            DT = {}
            for j in dict3:
                if j in dict1 and j in dict2:
                    DT.update({j:d1[i][j] + d2[i][j]})
                elif j in dict1:
                    DT.update({j:d1[i][j]})
                else:
                    DT.update({j:d2[i][j]})
            d3.update({i:DT})
        if type(d1[i]) == list:
            d3.update({i:d1[i] + d2[i]})
    else:
        if i in d1:
            d3.update({i:d1[i]})
        if i in d2:
            d3.update({i:d2[i]})
print (d3)            
        
```

```python
#TASK 3

d = {'a': 1, 'b': 4, 'c': 4, 'd': 2, 'e': 5}
N = 2
value = []
for i in d.values():
    if i not in value:
        value.append(i)
# print (value)        
largest_values = []
for i in range(N):
    max_val = None
    for v in value:
        if max_val == None or v > max_val:
            max_val = v
    largest_values.append(max_val)
    value.remove(max_val)
# print (largest_values)    
# print (value)    
cutoff = largest_values[-1]
# print( cutoff)
result = {}
for j in d:
    if d[j] >= cutoff:
        result[j] = d[j]
print(result)
```
